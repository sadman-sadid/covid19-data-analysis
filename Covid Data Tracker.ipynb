# Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading Datasets

```python
df = pd.read_csv("owid-covid-data.csv")
```

```python
print(df.head())
```

# Observing the Datatypes

```python
print(df.info())
```

# Creating a new DataFrame with Important Columns

```python
coln = ['location', 'date', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'reproduction_rate']

covid_df = df[coln]
```

# Converting Datatype of Date Column

```python
df['date'] = pd.to_datetime(df['date'])
df.info()
```

# Finding the null Values and replacing them 

```python
print(covid_df.isnull().sum)
```

```python
covid_df['reproduction_rate'] = covid_df['reproduction_rate'].fillna(0)
```

```python
print(covid_df.isnull().sum)
```

# Filtering Data for specific Countries(Bangladesh,India,Pakistan)

```python
BD_df = covid_df[covid_df['location'] == 'Bangladesh']
IN_df = covid_df[covid_df['location'] == 'India']
PK_df = covid_df[covid_df['location'] == 'Pakistan']
```

```python
print(BD_df.head(20))
```

```python
print(IN_df.head(20))
```

```python
print(PK_df.head(20))
```

# Lineplot of Total Cases over time in Bangladesh

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=BD_df,x='date',y='total_cases',label='Total Cases')
plt.title('COVID-19 Total Cases Over Time in Bangladesh')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

# Lineplot of Total Cases over time in India

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=IN_df,x='date',y='total_cases',label='Total Cases')
plt.title('COVID-19 Total Cases Over Time in India')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

# Lineplot of Total Cases over time in Pakistan

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=PK_df,x='date',y='total_cases',label='Total Cases')
plt.title('COVID-19 Total Cases Over Time in Pakistan')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

# Comparison of Total Cases between 3 Countries

```python
countries = ['Bangladesh','India','Pakistan']
multi_df = covid_df[covid_df['location'].isin(countries)]
```

```python
plt.figure(figsize=(14,7))
sns.lineplot(data=multi_df, x='date',y='total_cases',hue='location')

plt.title("COVID-19 Total Cases Comparison")
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.show()
```
